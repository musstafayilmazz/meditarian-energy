```python
import numpy as np
from scipy.spatial.distance import cdist
```

```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in km
    dlat = np.deg2rad(lat2 - lat1)
    dlon = np.deg2rad(lon2 - lon1)
    a = np.sin(dlat/2) * np.sin(dlat/2) + np.cos(np.deg2rad(lat1)) \
        * np.cos(np.deg2rad(lat2)) * np.sin(dlon/2) * np.sin(dlon/2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c
```

```python
seaports = df[['Latitude', 'Longitude']].values
```

```python
distances = cdist(df_filtered[['latitude', 'longitude']].values, seaports, 
                  lambda u, v: haversine(u[0], u[1], v[0], v[1]))
```

```python
nearest_seaport_index = np.argmin(distances, axis=1)
```

```python
distance_to_nearest_seaport = distances[np.arange(len(distances)), nearest_seaport_index]
```

```python
df_filtered['distance_to_nearest_seaport'] = distance_to_nearest_seaport
```
